3. **Nettoyage des données**

Le nettoyage des données est une étape essentielle pour garantir la qualité de l’analyse.

Dans ce dataset, certaines variables contiennent des valeurs égales à 0, ce qui n’est pas réaliste médicalement pour certaines mesures comme :

*le glucose
la pression sanguine
le BMI
l'insuline
l'épaisseur de la peau*

Ces valeurs sont donc considérées comme des valeurs manquantes.

Elles sont remplacées par la médiane de la variable, car la médiane est moins sensible aux valeurs extrêmes que la moyenne.

Cette étape permet d’obtenir un dataset plus fiable et prêt pour l’analyse exploratoire.

4. **Analyse exploratoire des données (EDA) — Interprétation du graphique**

Les histogrammes permettent d'observer la distribution des différentes variables du dataset et de comprendre les caractéristiques générales des patients.

Pregnancies
La majorité des patientes ont entre 0 et 3 grossesses, tandis que les valeurs élevées (plus de 10 grossesses) sont rares. Cela montre que la plupart des femmes de l'échantillon ont eu un nombre relativement faible de grossesses.

Glucose
Le niveau de glucose est principalement concentré entre 90 et 140. On observe aussi quelques valeurs élevées proches de 200, ce qui peut indiquer des patients ayant un risque plus élevé de diabète. La distribution montre que le glucose varie fortement entre les patients.

BloodPressure
La pression sanguine se situe majoritairement entre 60 et 80, ce qui correspond à des valeurs relativement normales. Les valeurs extrêmes sont peu fréquentes, ce qui indique que la majorité des patients ont une pression sanguine dans une plage standard.

SkinThickness
La plupart des valeurs se situent entre 20 et 40, ce qui indique que l'épaisseur de la peau varie peu dans la population étudiée. Cependant, certaines valeurs plus faibles ou plus élevées existent, ce qui peut refléter des différences physiques entre les patients.

Insulin
La distribution de l'insuline est très asymétrique (skewed). La majorité des patients ont des valeurs relativement faibles, mais quelques patients présentent des niveaux très élevés. Cela peut indiquer la présence de valeurs extrêmes (outliers).

BMI (Body Mass Index)
La majorité des patients ont un BMI entre 25 et 35, ce qui correspond à une population majoritairement en surpoids ou obèse. Cela est cohérent avec le fait que l'obésité est un facteur de risque important du diabète.

DiabetesPedigreeFunction
Cette variable mesure la prédisposition génétique au diabète. La plupart des patients ont des valeurs faibles (moins de 0.5), ce qui signifie que la majorité n'a pas une forte prédisposition génétique.

Age
La majorité des patients ont entre 20 et 40 ans. Les patients plus âgés sont moins nombreux dans le dataset, ce qui montre que l'échantillon est composé principalement d'adultes jeunes ou d'âge moyen.

Outcome
La variable Outcome montre deux groupes :

0 : non diabétiques
1 : diabétiques

On observe que les patients non diabétiques sont plus nombreux que les diabétiques, ce qui indique un déséquilibre léger dans les classes.

***A. Distribution du niveau de glucose***

La majorité des patients ont un niveau de glucose entre 90 et 130. Cependant, certaines valeurs dépassent 160, ce qui peut indiquer un risque plus élevé de diabète. La distribution montre que le glucose varie fortement entre les patients.

***B. Distribution du BMI***

La plupart des patients ont un BMI entre 25 et 35, ce qui correspond à une population majoritairement en surpoids. Quelques valeurs élevées (plus de 45) indiquent des cas d'obésité plus importante.

***C. Répartition des patients**

On observe que les patients non diabétiques sont plus nombreux que les diabétiques. Cela signifie que le dataset est légèrement déséquilibré, avec plus de cas négatifs que positifs.

***D. Glucose selon le diagnostic***

Les patients diabétiques ont généralement un niveau de glucose plus élevé que les patients non diabétiques. Cela montre que le glucose est un facteur important dans le diagnostic du diabète.

***E. BMI selon le diagnostic***

Le BMI médian des patients diabétiques est plus élevé que celui des non diabétiques. Cela suggère que le surpoids peut être associé à un risque plus élevé de diabète.

**6. Analyse de corrélation**

La matrice de corrélation montre la relation entre les différentes variables du dataset.

On observe que :

Glucose et Outcome (0.49) ont la corrélation la plus élevée, ce qui signifie que le niveau de glucose est fortement lié au diabète.
BMI et Outcome (0.31) montrent aussi une corrélation modérée, indiquant que un BMI élevé peut augmenter le risque de diabète.
Age et Outcome (0.24) suggèrent que le risque de diabète peut augmenter avec l’âge.
Pregnancies et Outcome (0.22) montrent une relation faible mais existante.

On remarque aussi certaines relations entre variables :

BMI et SkinThickness (0.54) ont une corrélation relativement forte.
Age et Pregnancies (0.54) sont également corrélés.

*Globalement, cette analyse montre que le glucose, le BMI et l’âge sont parmi les facteurs les plus associés au diabète dans ce dataset.*

**7. Formulation des hypothèses**
Pour le glucose et le BMI, l’hypothèse nulle affirme qu’il n’y a pas de différence entre diabétiques et non diabétiques, tandis que l’hypothèse alternative suppose une différence significative.
 Ces hypothèses permettent de vérifier si ces variables sont réellement liées au diabète.

**8. Test statistique (t-test)**
Le test t montre que les p-values pour le glucose et le BMI sont inférieures à 0.05. On rejette l’hypothèse nulle et on conclut que le glucose et le BMI influencent significativement le diabète.

**9. Comparaison des moyennes**
Les diabétiques présentent un glucose moyen et un BMI moyen plus élevés que les non diabétiques.
Cela confirme que l’hyperglycémie et le surpoids sont des facteurs de risque majeurs.

**10. Visualisations**
Les boxplots montrent des distributions plus hautes chez les diabétiques, la heatmap révèle une forte corrélation entre le glucose et le diabète, et la courbe ROC affiche une AUC autour de 0.75–0.80.
 Les graphiques confirment visuellement que le glucose et le BMI sont des variables importantes pour distinguer les diabétiques des non diabétiques.

**11. Modèle de Machine Learning (Régression Logistique)**
Le modèle atteint une accuracy de 70–80%, la matrice de confusion montre une bonne classification, et les coefficients indiquent que le glucose est le facteur le plus influent.
Interprétation : Le modèle IA valide les résultats statistiques et confirme que le glucose et le BMI sont des prédicteurs fiables du diabète.

**12. Conclusion**
Le glucose et le BMI sont significativement liés au diabète, et le glucose apparaît comme le facteur le plus déterminant.
Les résultats convergent vers la même conclusion : ces deux variables doivent être surveillées en priorité.

**13. Recommandations**
Il est conseillé de surveiller régulièrement le taux de glucose, de maintenir un poids équilibré, d’adopter une alimentation saine et de pratiquer une activité physique régulière.
 Ces mesures pratiques permettent de réduire le risque de diabète en contrôlant les deux facteurs principaux identifiés.
